```python
import csv
import numpy as np
import copy
```

```python
raw_database_dict = []

with open('data_231013_23.csv','r',newline='') as csvfile:
    csv_reader = csv.reader(csvfile)

    csv_header = next(csv_reader)    
    header_len = len(csv_header)
    
    

    for row in csv_reader:
        rawdata_dict={}
        save=True
        for idx, col in enumerate(row):
            # 날짜 데이터 string ->
            if len(col)==0:
                save=False
            rawdata_dict[csv_header[idx]] = col
        if save==True:
            raw_database_dict.append(rawdata_dict)

for dict in raw_database_dict:
    if "기온(°C)" in csv_header: 
        dict['기온(°C)']=float(dict['기온(°C)'])
    if "습도(%)" in csv_header:
        dict['습도(%)']=int(dict['습도(%)'])
    if "풍속(m/s)" in csv_header:
        dict['풍속(m/s)']=float(dict['풍속(m/s)'])
    if "현지기압(hPa)" in csv_header:
        dict['현지기압(hPa)']=float(dict["현지기압(hPa)"])
    if "전운량(10분위)" in csv_header:
        dict['전운량(10분위)']=float(dict["전운량(10분위)"])
    if "시정(10m)" in csv_header:
        dict['시정(10m)']=float(dict["시정(10m)"])

    

    # 날짜 수정
    time_info_string=dict["일시"]
    date_time = time_info_string.split(' ')

    dict["date"] = date_time[0]
    date = date_time[0].split('-')
    dict["year"] = int(date[0])
    dict["month"] = int(date[1])
    dict["day"] = int(date[2])

    dict["time"] = date_time[1]


print(raw_database_dict)
```

### Calculate W(temporal hour weight)

```python
sunriseList = []
sunsetList = []

with open('sunrise.txt',"r", encoding="utf-8") as sunriseText:
    riseText=sunriseText.read()
    riseMonths = riseText.split('/')
    for riseDays in riseMonths:
        sunriseList.append(riseDays.split('\n'))

with open('sunset.txt',"r", encoding="utf-8") as sunsetText:
    setText=sunsetText.read()
    setMonths = setText.split('/')
    for setDays in setMonths:
        sunsetList.append(setDays.split('\n'))



W_daytime = [0.05, 0.1, 0.51, 0.75, 0.95, 1.0, 0.9, 0.8, 0.59, 0.32, 0.22, 0.1]
W_dawn = [0.06, 0.08, 0.07, 0.11, 0.11, 0.11, 0.11, 0.13, 0.13, 0.13] # 일출 전 -1, -2, -3, -4시간 ~ AM 12:00
W_even = [0.08, 0.13, 0.13, 0.13, 0.13, 0.13, 0.13, 0.13, 0.13, 0.11] # 일몰 후 +1, +2시간 이후 ~ PM:11:59

'''
    func time2minute(params: time)
        "HH:MM"타입의 시간을 minute으로 변환하는 함수
'''
def time2minute(time: str):
    minute=0
    hour = int(time[0:2])
    minute = hour*60 + int(time[-2:])
    return minute
    
'''
   func time2minute(params: month, day, time)
        해당일의 월,일,시간의 temporal Hour Weight W 값을 구한다.
'''
def calcTempWeight(month: int, day: int, time: str):
    sunriseTime=sunriseList[month-1][day-1]
    sunsetTime=sunsetList[month-1][day-1]
    # print(f'일출시간:{sunriseList[month-1][day-1]}, 일몰시간:{sunsetList[month-1][day-1]}')
    dayMinute = time2minute(sunsetTime)-time2minute(sunriseTime)
    nowMinute = time2minute(time)-time2minute(sunriseTime)
    time_idx=round(nowMinute/dayMinute*11)

    W=0
    # 일몰 후
    if time_idx >= 12:
        W=W_even[time_idx-12]
    # 일출 전
    elif time_idx < 0:
        W=W_dawn[0-time_idx]
    else:
        W=W_daytime[time_idx]
    return W


'''
   func calcCn2(params: month, day, time)
        해당일의 월,일,시간의 Cn2 값을 구한다.
'''
def calcCn2(W, T: float, Ws: float, RH):
    f_T = np.float64(2*10e-15)*(273.15+T)
    f_U = np.float64(-2.5*10e-15)*Ws + np.float64(1.2*10e-15)*Ws*Ws + np.float64(-8.5*10e-17)*Ws*Ws*Ws
    RH=RH/100
    f_RH = np.float64(-2.8*10e-15)*RH + np.float64(2.9*10e-17)*RH*RH + np.float64(-1.1*10e-19)*RH*RH*RH
    Cn2 = np.float64(3.8*10e-14)*W + f_T + f_U + f_RH + np.float64(-5.3*10e-13)
    if Cn2 < 0:
        return np.float64(1e-15)
    return Cn2
    
    
```

```python
cn2_databse_dict=copy.deepcopy(raw_database_dict)
for dict in cn2_databse_dict:
    TemporalW = calcTempWeight(dict["month"], dict["day"], dict["time"]) 
    dict["cn2"] = calcCn2(TemporalW, dict["기온(°C)"],  dict["풍속(m/s)"], dict["습도(%)"])
    print(f'{dict["cn2"]} from W: {TemporalW}, T:{dict["기온(°C)"]}, U:{dict["풍속(m/s)"]}, H:{dict["습도(%)"]}')
```
